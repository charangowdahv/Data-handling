```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
print("step 1")
```

```python
fig=plt.figure()
ax=plt.axes()
ax.grid()
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,1000)
print(x)
ax.plot(x,np.sin(x));
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,1000)
print(x)
ax.plot(x,np.cos(x))
ax.plot(x,np.sin(x))
ax.plot(x,np.tan(x));
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,1000)
print(x)
ax.plot(x,np.sin(x))
ax.plot(x,np.cos(x))
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel');
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,1000)
print(x)
ax.plot(x,np.sin(x),label='sin');
ax.plot(x,np.cos(x),label='cos');
ax.set_title('multiple line')
ax.set_xlabel('period')
ax.set_ylabel('amplitude')
ax.legend()
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,1000)
print(x)
ax.plot(x,np.sin(x),label='sin',color='black');
ax.plot(x,np.cos(x),label='cos',color='green');
ax.set_title('multiple line')
ax.set_xlabel('period')
ax.set_ylabel('amplitude')
ax.legend()
plt.show()

```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,1000)
print(x)
ax.plot(x,np.sin(x),label='sin',color='black');
ax.plot(x,np.cos(x),label='cos',color='green',linestyle='dashdot');
ax.set_title('multiple line')
ax.set_xlabel('period')
ax.set_ylabel('amplitude')
ax.legend()
plt.show()
```
